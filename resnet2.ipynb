```python
# Generate augmented images
datagen = ImageDataGenerator(
    height_shift_range=0.2,
    width_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    rotation_range=40,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest',
)

datagen.fit(x_train)
```

```python
# see example augmentation images
for X_batch, y_batch in datagen.flow(x_train, y_train, batch_size=9):
    for i in range(0, 9):
        plt.subplot(330 + 1 + i)
        plt.imshow(X_batch[i].astype(np.uint8))
    plt.show()
    break
```

```python
base_model = ResNet50(include_top=False,weights='imagenet',input_shape=(32,32,3),classes=100)
```

```python
model_1= models.Sequential()
model_1.add(base_model)
 
model_1.add(layers.Activation('relu'))
model_1.add(layers.Flatten())
model_1.add(layers.Dropout(0.5))
model_1.add(layers.Dense(100, activation='softmax', kernel_initializer='he_normal'))
```

```python
model_1.compile(optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy'])
```

```python
model_1.summary()
```
